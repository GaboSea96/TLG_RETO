```python
import pandas as pd
# a. leemos los datos y convertimos ["date"] en datetime64

country_vaccinations = pd.read_csv('/content/drive/MyDrive/country_vaccinations.csv', parse_dates = ['date'])

# b. mostramos la estructura y tipo de datos.
country_vaccinations.info()
```

```python
#c. Detemrinar la cantidad de vacunas aplicadas de cada compañía
suma_vaccines = country_vaccinations[['vaccines','daily_vaccinations']].groupby('vaccines').sum()
print(suma_vaccines)
```

```python
#d.Obtener la cantidad de vacunas aplicadas en todo el mundo
vacunas_aplicadas = country_vaccinations['daily_vaccinations'].sum()
print('el número total de vacunas aplicadas en el mundo es de', vacunas_aplicadas)
```

```python
#e.calcular el promedio de vacunas aplicadas por pais
promedio_vacunas_por_pais = country_vaccinations[['country','daily_vaccinations']].groupby('country').mean()
print(promedio_vacunas_por_pais)
```

```python
#f. Determinar la cantidad de vacunas aplicadas el dia 29/01/21 en todo el mundo
vacunas_por_fecha = country_vaccinations[country_vaccinations['date']=='2021-01-21']
suma_vacunas_por_fecha = vacunas_por_fecha['daily_vaccinations'].sum()
print('se pusieron un total de',suma_vacunas_por_fecha,'el dia 29/01/21 en todo el mundo')
```

```python
# g. Crear un DataFrame nuevo denominado conDiferencias que contenga los datos originales y
# una columna derivada(diferencias) con las diferencias de aplicación entre las columnas
# daily_vaccinations y daily_vaccinations_raw
conDiferencias = pd.read_csv('/content/drive/MyDrive/country_vaccinations.csv', parse_dates = ['date'])
conDiferencias['diferencias']= conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias.head()
```

```python
#h. obtener el periodo de tiempo entre el registro con fecha mas reciente y el registro con fecha mas antigua
periodo = country_vaccinations['date'].max() - country_vaccinations['date'].min()
print('el preiodo de tiempo entre ambas fechas es de', periodo)
```

```python
#i. Crear un DataFrame nuevo denominado conCantidad que contenga los datos originales y una columna derivda(canVac)
# con la cantidad de vacunas utilizadas cada dia.

conCantidad = pd.read_csv('/content/drive/MyDrive/country_vaccinations.csv', parse_dates = ['date'])
conCantidad['canVac'] = conCantidad['vaccines'].str.split(', ').str.len()
conCantidad.head()
```

```python
#j. Generar un DataFrame denomionado antes20 con todos los registros que se hayan realizado antes del
#20 de diciembre de 2020

datos_vacunas = pd.read_csv('/content/drive/MyDrive/country_vaccinations.csv', parse_dates = ['date'])
fecha = pd.datetime(2020,12,20)
antes20 = datos_vacunas[datos_vacunas['date'] > fecha]
print('la fecha minima es' ,antes20['date'].min())
antes20.head()
```

```python
#k.Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna pfizer
pfizer = country_vaccinations[country_vaccinations['vaccines'].str.contains('Pfizer', case = False)]
pfizer.head()
```

```python
#Almacenar los resultados generados (conDiferencias,conCantidad,antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx..
with pd.ExcelWriter('/content/drive/MyDrive/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='Diferencias')
    conCantidad.to_excel(writer,sheet_name='Cantidad')
    antes20.to_excel(writer,sheet_name="antes20")
    pfizer.to_excel(writer,sheet_name='pfizer')

```
